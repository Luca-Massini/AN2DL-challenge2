```
from PIL import Image

class CustomDataset(tf.keras.utils.Sequence):

  """
    CustomDataset inheriting from tf.keras.utils.Sequence.

    3 main methods:
      - __init__: save dataset params like directory, filenames..
      - __len__: return the total number of samples in the dataset
      - __getitem__: return a sample from the dataset

    Note: 
      - the custom dataset return a single sample from the dataset. Then, we use 
        a tf.data.Dataset object to group samples into batches.
      - in this case we have a different structure of the dataset in memory. 
        We have all the images in the same folder and the training and validation splits
        are defined in text files.

  """

  def __init__(self, dataset_dir, which_subset, img_generator=None, mask_generator=None, 
               preprocessing_function=None, out_shape=[512, 384],validation_part=0.2,projects=['Weedelec','Pead','Roseau','Bipbip'],plants=['Haricot','Mais']):
    self.png_or_jpg = []
    subset_filenames = []
    subset_direc = []
    for project in projects :
      for plant in plants :
        path=os.path.join(dataset_dir,project,plant)
        list_ = os.listdir(os.path.join(path,'Images')) # dir is your directory path
        number_files = len(list_)

        training_num_max = int(number_files - int(number_files *  validation_part))
        print("training files: ", training_num_max, "validation_part: ", number_files - training_num_max)
        if which_subset == 'training':
          number_max = training_num_max
          for ii in range(0,number_max):
            subset_filenames.append(os.path.splitext(list_[ii])[0])
            subset_direc.append(path)
            if project!='Roseau':
              self.png_or_jpg.append(1)
            else:
              self.png_or_jpg.append(0)
        elif which_subset == 'validation':
          number_max = number_files - training_num_max
          for ii in range(number_files - 1, number_files - number_max, -1):
            subset_filenames.append(os.path.splitext(list_[ii])[0])
            subset_direc.append(path)
            if project!='Roseau':
              self.png_or_jpg.append(1)
            else:
              self.png_or_jpg.append(0)


    self.which_subset = which_subset
    self.dataset_dir = dataset_dir
    self.subset_direc = subset_direc
    self.subset_filenames = subset_filenames
    self.img_generator = img_generator
    self.mask_generator = mask_generator
    self.preprocessing_function = preprocessing_function
    self.out_shape = out_shape

  def __len__(self):
    return len(self.subset_filenames)

  def read_rgb_mask(img_path):
    '''
    img_path: path to the mask file
    Returns the numpy array containing target values
    '''

    mask_img = Image.open(img_path)
    mask_arr = np.array(mask_img)

    new_mask_arr = np.zeros(mask_arr.shape[:2], dtype=mask_arr.dtype)

    # Use RGB dictionary in 'RGBtoTarget.txt' to convert RGB to target
    new_mask_arr[np.where(np.all(mask_arr == [216, 124, 18], axis=-1))] = 0
    new_mask_arr[np.where(np.all(mask_arr == [255, 255, 255], axis=-1))] = 1
    new_mask_arr[np.where(np.all(mask_arr == [216, 67, 82], axis=-1))] = 2

    return new_mask_arr
  
  def __getitem__(self, index):
    # Read Image
    curr_filename = self.subset_filenames[index]
    curr_dir = self.subset_direc[index]
    if self.png_or_jpg[index] != 0:
      img = Image.open(os.path.join(curr_dir, 'Images', curr_filename + '.jpg'))
    else:
      img = Image.open(os.path.join(curr_dir, 'Images', curr_filename + '.png'))
    mask = Image.open(os.path.join(curr_dir, 'Masks', curr_filename + '.png'))

    # Resize image 
    img = img.resize(self.out_shape)
    mask = mask.resize(self.out_shape, resample=Image.NEAREST)

    mask_arr = np.array(mask)
    new_mask_arr = np.zeros(mask_arr.shape[:2], dtype=mask_arr.dtype)

    # Use RGB dictionary in 'RGBtoTarget.txt' to convert RGB to target
    new_mask_arr[np.where(np.all(mask_arr == [216, 124, 18], axis=-1))] = 0
    new_mask_arr[np.where(np.all(mask_arr == [255, 255, 255], axis=-1))] = 1
    new_mask_arr[np.where(np.all(mask_arr == [216, 67, 82], axis=-1))] = 2
    
    img_arr = np.array(img)
    mask_arr = new_mask_arr

    # in this dataset 255 mask label is assigned to an additional class, which corresponds 
    # to the contours of the objects. We remove it for simplicity.
    #mask_arr[mask_arr == 255] = 0  

    mask_arr = np.expand_dims(mask_arr, -1)

    if self.which_subset == 'training':
      if self.img_generator is not None and self.mask_generator is not None:
        # Perform data augmentation
        # We can get a random transformation from the ImageDataGenerator using get_random_transform
        # and we can apply it to the image using apply_transform
        img_t = self.img_generator.get_random_transform(img_arr.shape, seed=SEED)
        mask_t = self.mask_generator.get_random_transform(mask_arr.shape, seed=SEED)
        img_arr = self.img_generator.apply_transform(img_arr, img_t)
        # ImageDataGenerator use bilinear interpolation for augmenting the images.
        # Thus, when applied to the masks it will output 'interpolated classes', which
        # is an unwanted behaviour. As a trick, we can transform each class mask 
        # separately and then we can cast to integer values (as in the binary segmentation notebook).
        # Finally, we merge the augmented binary masks to obtain the final segmentation mask.
        out_mask = np.zeros_like(mask_arr)
        for c in np.unique(mask_arr):
          if c > 0:
            curr_class_arr = np.float32(mask_arr == c)
            curr_class_arr = self.mask_generator.apply_transform(curr_class_arr, mask_t)
            # from [0, 1] to {0, 1}
            curr_class_arr = np.uint8(curr_class_arr)
            # recover original class
            curr_class_arr = curr_class_arr * c 
            out_mask += curr_class_arr
    else:
      out_mask = mask_arr
    
    if self.preprocessing_function is not None:
        img_arr = self.preprocessing_function(img_arr)

    return img_arr, np.float32(out_mask)
```

```

def create_model(start_f):
  input_shape = [img_h, img_w, 3]
  inputs = tf.keras.Input(shape=(img_h, img_w, 3))
  i_1 = tf.keras.layers.Conv2D(filters = start_f, 
                                kernel_size = (2,2),
                                strides=(1, 1),
                                padding='same')(inputs)
  i_2 = tf.keras.layers.Conv2D(filters = start_f, 
                                kernel_size = (2,2),
                                strides=(1, 1),
                                padding='same')(i_1)
  x_0 = tf.keras.layers.Conv2D(filters = start_f, 
                                kernel_size = (3,3),
                                strides=(1, 1),
                                padding='same')(inputs)
  x_1 = tf.keras.layers.Conv2D(filters = start_f, 
                                kernel_size = (3,3),
                                strides=(1, 1),
                                padding='same')(x_0)
  i_3 = tf.keras.layers.Average()([i_2, x_1])
  input_shape = None


  start_f *= 2
  x_2 = tf.keras.layers.BatchNormalization(axis=-1)(i_3)
  x_3  = tf.keras.layers.Activation('relu')(x_2)
  x_4 = tf.keras.layers.MaxPool2D(pool_size=(2, 2))(x_3)
  p_1 = tf.keras.layers.Conv2D(filters=start_f, 
                      kernel_size=(2, 2),
                      strides = (1,1),
                      padding = 'same')(x_4)
  p_2 = tf.keras.layers.Conv2D(filters=start_f, 
                      kernel_size=(2, 2),
                      strides = (1,1),
                      padding = 'same')(p_1)
  c_1 = tf.keras.layers.Conv2D(filters=start_f, 
                      kernel_size=(3, 3),
                      strides = (1,1),
                      padding = 'same')(x_4)
  x_5 = tf.keras.layers.Conv2D(filters=start_f, 
                      kernel_size=(3, 3),
                      strides = (1,1),
                      padding = 'same')(c_1)
  u_1 = tf.keras.layers.Average()([p_2, x_5])
  x_6 = tf.keras.layers.BatchNormalization(axis=-1)(u_1)
  x_7 = tf.keras.layers.Activation('relu')(x_6)
  x_8 = tf.keras.layers.MaxPool2D(pool_size=(2, 2))(x_7)



  start_f *= 2
  p_3 = tf.keras.layers.Conv2D(filters=start_f, 
                      kernel_size=(2, 2),
                      strides = (1,1),
                      padding = 'same')(x_8)
  p_4 = tf.keras.layers.Conv2D(filters=start_f, 
                      kernel_size=(2, 2),
                      strides = (1,1),
                      padding = 'same')(p_3)
  c_2 = tf.keras.layers.Conv2D(filters=start_f, 
                      kernel_size=(3, 3),
                      strides = (1,1),
                      padding = 'same')(x_8)
  x_9 = tf.keras.layers.Conv2D(filters=start_f, 
                      kernel_size=(3, 3),
                      strides = (1,1),
                      padding = 'same')(c_2)
  u_2 = tf.keras.layers.Average()([p_4, x_9])
  y_1 = tf.keras.layers.BatchNormalization(axis=-1)(u_2)
  y_2 = tf.keras.layers.Activation('relu')(y_1)
  y_3 = tf.keras.layers.MaxPool2D(pool_size=(2, 2))(y_2)




  start_f *= 2
  p_5 = tf.keras.layers.Conv2D(filters=start_f, 
                      kernel_size=(2, 2),
                      strides = (1,1),
                      padding = 'same')(y_3)
  p_6 = tf.keras.layers.Conv2D(filters=start_f, 
                      kernel_size=(2, 2),
                      strides = (1,1),
                      padding = 'same')(p_5)
  c_3 = tf.keras.layers.Conv2D(filters=start_f, 
                      kernel_size=(3, 3),
                      strides = (1,1),
                      padding = 'same')(y_3)
  y_4 = tf.keras.layers.Conv2D(filters=start_f, 
                      kernel_size=(3, 3),
                      strides = (1,1),
                      padding = 'same')(c_3)
  u_3 = tf.keras.layers.Average()([p_6, y_4])
  y_5 = tf.keras.layers.BatchNormalization(axis=-1)(u_3)
  y_6 = tf.keras.layers.Activation('relu')(y_5)
  y_7 = tf.keras.layers.MaxPool2D(pool_size=(2, 2))(y_6)





  start_f *= 2
  p_7 = tf.keras.layers.Conv2D(filters=start_f, 
                      kernel_size=(2, 2),
                      strides = (1,1),
                      padding = 'same')(y_7)
  p_8 = tf.keras.layers.Conv2D(filters=start_f, 
                      kernel_size=(2, 2),
                      strides = (1,1),
                      padding = 'same')(p_7)
  c_4 = tf.keras.layers.Conv2D(filters=start_f, 
                      kernel_size=(3, 3),
                      strides = (1,1),
                      padding = 'same')(y_7)
  y_8 = tf.keras.layers.Conv2D(filters=start_f, 
                      kernel_size=(3, 3),
                      strides = (1,1),
                      padding = 'same')(c_4)
  u_4 = tf.keras.layers.Average()([y_8, p_8])
  y_9 = tf.keras.layers.BatchNormalization(axis=-1)(u_4)
  z_1 = tf.keras.layers.Activation('relu')(y_9)



  #MIDDLE BLOCK
  m_1 = tf.keras.layers.Conv2D(filters=start_f, 
                      kernel_size=(3, 3),
                      strides = (1,1),
                      padding = 'same')(z_1)
  m_2 = tf.keras.layers.Conv2D(filters=start_f, 
                      kernel_size=(3, 3),
                      strides = (1,1),
                      padding = 'same')(m_1)
  m_3 = tf.keras.layers.BatchNormalization(axis=-1)(m_2)
  m_4 = tf.keras.layers.Activation('relu')(m_3)





  #DECODER
  start_f = start_f // 2
  g_1 = tf.keras.layers.Conv2DTranspose(filters=start_f, kernel_size=(2,2), strides=(2, 2), padding='same')(m_4)
  g_2 = tf.keras.layers.Conv2DTranspose(filters=start_f, kernel_size=(3,3), strides=(1, 1), padding='same')(g_1)
  d_2 = tf.keras.layers.Conv2DTranspose(filters=start_f, kernel_size=(3,3), strides=(2, 2), padding='same')(m_4)
  d_3 = tf.keras.layers.Conv2DTranspose(filters=start_f, kernel_size=(3,3), strides=(1,1), padding='same')(d_2)
  u_5 = tf.keras.layers.Average()([g_2, d_3])
  d_4 = tf.keras.layers.ReLU()(u_5)
  d_6 = tf.keras.layers.Add()([d_4,y_6])

  start_f = start_f // 2
  g_3 = tf.keras.layers.Conv2DTranspose(filters=start_f, kernel_size=(2,2), strides=(2, 2), padding='same')(d_6)
  g_4 = tf.keras.layers.Conv2DTranspose(filters=start_f, kernel_size=(2,2), strides=(1, 1), padding='same')(g_3)
  d_7 = tf.keras.layers.Conv2DTranspose(filters=start_f, kernel_size=(3,3), strides=(2, 2), padding='same')(d_6)
  d_8 = tf.keras.layers.Conv2DTranspose(filters=start_f, kernel_size=(3,3), strides=(1, 1), padding='same')(d_7)
  u_6 = tf.keras.layers.Average()([d_8, g_4])
  d_9 = tf.keras.layers.ReLU()(u_6)
  e_1 = tf.keras.layers.Add()([d_9,y_2])


  start_f = start_f // 2
  g_5 = tf.keras.layers.Conv2DTranspose(filters=start_f, kernel_size=(2,2), strides=(2, 2), padding='same')(e_1)
  g_6 = tf.keras.layers.Conv2DTranspose(filters=start_f, kernel_size=(2,2), strides=(1, 1), padding='same')(g_5)
  e_2 = tf.keras.layers.Conv2DTranspose(filters=start_f, kernel_size=(3,3), strides=(2, 2), padding='same')(e_1)
  e_3 = tf.keras.layers.Conv2DTranspose(filters=start_f, kernel_size=(3,3), strides=(1, 1), padding='same')(e_2)
  u_7 = tf.keras.layers.Average()([e_3, g_6])
  e_4 = tf.keras.layers.ReLU()(u_7)
  e_6 = tf.keras.layers.Add()([e_4,x_7])

  start_f = start_f // 2
  g_7 = tf.keras.layers.Conv2DTranspose(filters=start_f, kernel_size=(2,2), strides=(2, 2), padding='same')(e_6)
  g_8 = tf.keras.layers.Conv2DTranspose(filters=start_f, kernel_size=(2,2), strides=(1, 1), padding='same')(g_7)
  e_8 = tf.keras.layers.Conv2DTranspose(filters=start_f, kernel_size=(3,3), strides=(2, 2), padding='same')(e_6)
  e_9 = tf.keras.layers.Conv2DTranspose(filters=start_f, kernel_size=(3,3), strides=(1,1), padding='same')(e_8)
  u_8 = tf.keras.layers.Average()([g_8, e_9])
  f_1 = tf.keras.layers.ReLU()(u_8)
  f_2 = tf.keras.layers.Add()([f_1,x_3])
  h = tf.keras.layers.Dropout(0.4, seed=SEED)(f_2)


  
  last = tf.keras.layers.Conv2D(filters=3,
                                kernel_size=(1, 1),
                                strides=(1, 1),
                                padding='same',
                                activation='softmax')(h)
  model_1 = tf.keras.Model(inputs=inputs, outputs=last)
  print("first model summary")
  model_1.summary()

  return model_1

  

```

```
model=create_model(start_f=16)
```

```
# Optimization params
# -------------------

# Loss
# Sparse Categorical Crossentropy to use integers (mask) instead of one-hot encoded labels
loss = tf.keras.losses.SparseCategoricalCrossentropy() 
# learning rate
lr = 7e-5
optimizer = tf.keras.optimizers.Adam(learning_rate=lr)
# -------------------

# Here we define the intersection over union for each class in the batch.
# Then we compute the final iou as the mean over classes
def meanIoU(y_true, y_pred):
    # get predicted class from softmax
    y_pred = tf.expand_dims(tf.argmax(y_pred, -1), -1)

    per_class_iou = []

    for i in range(1,3): # exclude the background class 0
      # Get prediction and target related to only a single class (i)
      class_pred = tf.cast(tf.where(y_pred == i, 1, 0), tf.float32)
      class_true = tf.cast(tf.where(y_true == i, 1, 0), tf.float32)
      intersection = tf.reduce_sum(class_true * class_pred)
      union = tf.reduce_sum(class_true) + tf.reduce_sum(class_pred) - intersection
    
      iou = (intersection + 1e-7) / (union + 1e-7)
      per_class_iou.append(iou)

    return tf.reduce_mean(per_class_iou)

# Validation metrics
# ------------------
metrics = ['accuracy', meanIoU]
# ------------------

# Compile Model
model.compile(optimizer=optimizer, loss=loss, metrics=metrics)
```

```
import os
import json
import numpy as np
from PIL import Image


def rle_decode(rle, shape):
    s = rle.split()
    starts, lengths = [np.asarray(x, dtype=int) for x in (s[0:][::2], s[1:][::2])]
    starts -= 1
    ends = starts + lengths
    img = np.zeros(shape[0]*shape[1], dtype=np.uint8)
    for lo, hi in zip(starts, ends):
        img[lo:hi] = 1
    return img.reshape(shape)




# Just for completeness, here we show how the rle encoded mask
# is transformed back to a numpy array

# Read the submission.json file generated by the
# 'prepare_submission.py' script
with open('/content/drive/MyDrive/submission_prova.json', 'r') as f:
      submission_dict = json.load(f)

img_name = 'Weedelec_mais_2019-09-25T121103-128'
img_shape = submission_dict[img_name]['shape']

rle_encoded_crop = submission_dict[img_name]['segmentation']['crop']
rle_encoded_weed = submission_dict[img_name]['segmentation']['weed']

# Reconstruct crop and weed binary masks
crop_mask = rle_decode(rle_encoded_crop, shape=img_shape)
weed_mask = rle_decode(rle_encoded_weed, shape=img_shape)

# Reconstruct original mask
# weed_mask * 2 allows to convert ones into target 2 (weed label)
reconstructed_mask = crop_mask + (weed_mask * 2)

# Just for visualisation purposes, save RGB reconstructed mask
# Use again the dictionary in 'RGBtoTarget.txt'.
reconstructed_rgb_arr = np.zeros(shape=img_shape + [3])
reconstructed_rgb_arr[reconstructed_mask == 1] = [255, 255, 255]
reconstructed_rgb_arr[reconstructed_mask == 2] = [216, 67, 82]

reconstructed_rgb_img = Image.fromarray(
    np.uint8(reconstructed_rgb_arr))
reconstructed_rgb_img.save('/content/drive/MyDrive/rle_decoded_rgb_mask.png')

```
